```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import pandas as pd
import os
import wget
import time
import numpy as np
```

```python
options = webdriver.ChromeOptions() 
options.add_argument("start-maximized")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)

driver = webdriver.Chrome(options=options, executable_path=r'D:/notebook/chromedriver/chromedriver.exe')
driver.get('https://shopee.co.id/shop/54873248/search')
```

```python
links = driver.find_elements_by_tag_name('a')
links = [link.get_attribute('href') for link in links]

links = links[16:]
```

```python
harga_produk = []
for i in links:
    driver.get(i)
    try:
        harga = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "Ybrg9j"))).text
        harga_produk.append(harga)
    finally:
        driver.back()
```

```python
nama_produk = []
for i in links:
    driver.get(i)
    try:
        nama = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "attM6y"))).text
        nama_produk.append(nama)
    finally:
        driver.back()
```

```python
deskrip_produk = []
for i in links:
    driver.get(i)
    try:
        deskrip = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, '//*[@id="main"]/div/div[2]/div[2]/div[2]/div[2]/div[2]/div[1]/div[1]/div[2]/div[2]/div/span'))).text
        deskrip_produk.append(deskrip)
    finally:
        driver.back()
```

```python
stok_produk = []
for i in links:
    driver.get(i)
    try:
        stok = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, '#main > div > div._193wCc > div.page-product.page-product--mall > div.container > div.product-briefing.flex.card.zINA0e > div.flex.flex-auto._3-GQHh > div > div._2nr4HE > div > div.flex._3AHLrn._2XdAdB > div > div.flex.items-center._90fTvx > div.flex.items-center > div:nth-child(2)'))).text
        stok_produk.append(stok)
    finally:
        driver.back()
```

```python
gambar_produk = []
for i in links:
    driver.get(i)
    try:
        time.sleep(10)
        k = driver.find_elements_by_css_selector("div[class='_12uy03 _2GchKS']")
        jb = []
        for j in k:
            b = j.value_of_css_property("background-image")
            jb.append(b)
        gambar_produk.append(jb)
    finally:
        driver.back()
```

```python
hasil = {'nama produk': nama_produk,
        'harga': harga_produk,
         'stok_produk': stok_produk,
        'specs' : deskrip_produk,
        'gambar produk': gambar_produk}
```
